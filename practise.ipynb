```python
#The provided code stub will read in a dictionary containing key/value pairs of name:[marks] 
#for a list of students. Print the average of the marks array for the student name provided, showing 2 places after the decimal.

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
marks = [scores for name, scores in student_marks.items() if name == query_name]
marks = sum(marks, [])
x = (sum(marks))/3
print("%.2f" %x)
```

```python
#You are given a string and your task is to swap cases. 
#In other words, convert all lowercase letters to uppercase letters and vice versa.

def swap_case(s):
    result = ''
    
    for i in range(len(s)):
        if s[i].isupper():
            result += s[i].lower()
        elif s[i].islower():
            result += s[i].upper()
        else:
            result += s[i] 
    
    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
#You are asked to ensure that the first and last names of people begin with a capital letter in their passports. 
#For example, alison heck should be capitalised correctly as Alison Heck.
#Given a full name, your task is to capitalize the name appropriately.

def solve(s):
    s = s.split(" ")
    s = [i.title() if i.isalpha() else i.capitalize() for i in s]
    s = " ".join(s)
    return s

if __name__ == '__main__':
    s = input()
    result = solve(s)
    print(result)
```

```python
from itertools import product

def maximize_s(lists, M):
    max_val = 0
    for combination in product(*lists):
        total = sum(x**2 for x in combination) % M
        if total > max_val:
            max_val = total
    print(max_val)
    return

numbers, modulo = map(int, input("").split())
lister = []

while numbers:
    count, listed = input('').split(' ',1)
    lister.append(listed)
    numbers -= 1

lister = [x.split() for x in lister]
lister = [list(map(int, sublist)) for sublist in lister]

maximize_s(lister, modulo)
```

```python
empty = []
operations = []

def apply(N, empty):
    
    match N:

        case _ if "insert" in N:
            operation, index, value = N.split()
            index, value = int(index), int(value)
            empty.insert(index, value)

        case "print":
            print(empty)

        case _ if "remove" in N:
            operation, value = N.split()
            value = int(value)
            empty.remove(value)

        case _ if "append" in N:
            operation, value = N.split()
            value = int(value)
            empty.append(value)

        case "sort":
            empty.sort()

        case "pop":
            empty.pop()

        case "reverse":
            empty.reverse()

if __name__ == '__main__':
    N = int(input())

for _ in range(N):
    sub = input()
    operations.append(sub)

for i in operations:
    apply(i, empty)
```
