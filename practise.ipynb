```python
#Given the names and grades for each student in a class of N students
#store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

# Input number of students
n = int(input("Enter the number of students: "))

# Create the nested list for name and score
name_score = []
for _ in range(n):
    name = input("Enter the student's name: ")
    score = float(input("Enter the student's score: "))
    name_score.append([name, score])

# Step 1: Extract all unique scores and sort them
unique_scores = sorted(set([score for name, score in name_score]))
print(unique_scores)

# Step 2: Identify the second lowest score
second_lowest = unique_scores[1]
print(second_lowest)

# Step 3: Get names of students with the second lowest score
second_lowest_students = sorted([name for name, score in name_score if score == second_lowest])
print(second_lowest_students)

# Step 5: Print each name on a new line
for name in second_lowest_students:
    print(name)
```

```python
#The provided code stub will read in a dictionary containing key/value pairs of name:[marks] 
#for a list of students. Print the average of the marks array for the student name provided, showing 2 places after the decimal.

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    
marks = [scores for name, scores in student_marks.items() if name == query_name]
marks = sum(marks, [])
x = (sum(marks))/3
print("%.2f" %x)
```

```python
#You are given a string and your task is to swap cases. 
#In other words, convert all lowercase letters to uppercase letters and vice versa.

def swap_case(s):
    result = ''
    
    for i in range(len(s)):
        if s[i].isupper():
            result += s[i].lower()
        elif s[i].islower():
            result += s[i].upper()
        else:
            result += s[i] 
    
    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
#Mr. Vincent works in a door mat manufacturing company. One day, he designed a new door mat with the following specifications:
#Mat size must be NXM. (N is an odd natural number, and M is 3 times N.)
#The design should have 'WELCOME' written in the center.
#The design pattern should only use |, . and - characters.

# Enter your code here. Read input from STDIN. Print output to STDOUT
N, M = input().split()
N = int(N)
M = int(M)

for i in range(1,(N//2)+1):
    x = ".|." * ((i*2)-1)
    print(x.center(M, "-"))

print("WELCOME".center(M, "-"))

for i in range((N//2),0,-1):
    x = ".|." * ((i*2)-1)
    print(x.center(M, "-"))
```

```python
#You are asked to ensure that the first and last names of people begin with a capital letter in their passports. 
#For example, alison heck should be capitalised correctly as Alison Heck.
#Given a full name, your task is to capitalize the name appropriately.

def solve(s):
    s = s.split(" ")
    s = [i.title() if i.isalpha() else i.capitalize() for i in s]
    s = " ".join(s)
    return s

if __name__ == '__main__':
    s = input()
    result = solve(s)
    print(result)
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import Counter

numshoe = int(input(""))
shoesizes = Counter(list(map(int,input("").split()))[:numshoe])
numcust = int(input(""))
income = 0 

for _ in range(numcust):
    shoe, prize = map(int, input("").split())

    if shoesizes[shoe]>0:
        shoesizes[shoe] -= 1
        income += prize

    else:
        pass

print(income)
```
